## Travel Time

```python
con = sqlite3.connect('results0point1lambda.db')
cur = con.cursor()
cur.execute("SELECT AGENT, CAPACITY, MIN, Q1, MEDIAN, MEAN, Q3, MAX from capacity_runs_tratime")
travel_times = cur.fetchall()
cur.execute("SELECT AGENT, CAPACITY, MIN, Q1, MEDIAN, MEAN, Q3, MAX from capacity_runs_vottime")
vot_times = cur.fetchall()

agents = list(set([x[0] for x in travel_times]))
colours = ['#CC4F1B', '#1B2ACC', '#3F7F4C']
for c, agent in zip(colours, agents):
    results_tts = [x for x in travel_times if x[0] == agent]
    x_axis = [x[1] for x in results_tts]
    mins = [x[2] for x in results_tts]
    lwqs = [x[3] for x in results_tts]
    meds = [x[4] for x in results_tts]
    avgs = [x[5] for x in results_tts]
    upqs = [x[6] for x in results_tts]
    maxs = [x[7] for x in results_tts]
    
    plt.plot(x_axis, avgs, label=agent)
    plt.fill_between(x_axis, lwqs, upqs, alpha=0.2)

plt.title("Lambda = 0.1")
plt.ylabel("Travel Time")
plt.legend()
# plt.ylim(100,150)
plt.show()
```
